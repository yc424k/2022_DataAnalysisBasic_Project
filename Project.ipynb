```python
import pandas as pd
```

```python
df = pd.read_csv('CAR DETAILS FROM CAR DEKHO.csv')
# print(df.head())
print(df.info())
# 결측된 데이터가 없다는걸 알 수 있음
```

```python
df.index
```

```python
sorted(list(df['year'].unique()))
# 총 4340개의 차량이 있고, 1992년부터 2020년까지의 데이터가 있음
```

```python
# 년도별 차량갯수 출력
# df['year'].value_counts()
df['year'].value_counts().sort_index()
```

```python
#x축은 출고 년도, y축은 평균 판매가로 꺾은선 그래프를 그리기
df.groupby('year')['selling_price'].mean().plot.line()
```

```python
# x축은 자동차 브랜드, y축은 자동차 보유 대수로 막대그래프를 내림차순으로 그리기
df['name'].str[:4].value_counts().sort_values(ascending=False).plot.bar()
```

```python
# 출고년도와 가격의 연관성 그래프 그리기
df.groupby('year')['selling_price'].mean().plot.line()
```
